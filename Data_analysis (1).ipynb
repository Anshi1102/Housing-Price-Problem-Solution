```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
# https://www.kaggle.com/competitions/home-data-for-ml-course
df = pd.read_csv("train.csv")
df.shape
df.head()
```

```python
# Missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# Data types
df.dtypes
df.dtypes.value_counts()
```

```python
# Basic Stats
df.info()
```

```python
df.describe()
```

```python
sns.histplot(df['SalePrice'], kde=True)
plt.title('Sale Price Distribution')
```

```python
# Log transformation
df['LogSalePrice'] = np.log1p(df['SalePrice'])
sns.histplot(df['LogSalePrice'], kde=True)
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr[['SalePrice']].sort_values(by='SalePrice', ascending=False), annot=True, cmap='coolwarm')
```

```python
df['TotalBath'] = df['FullBath'] + 0.5*df['HalfBath']
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['Age'] = df['YrSold'] - df['YearBuilt']
df['RemodelAge'] = df['YrSold'] - df['YearRemodAdd']
```

```python
sns.scatterplot(x='TotalSF', y='SalePrice', data=df)
```

```python
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
```

```python
# Fill NA with 0 for numerical features
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill NA with mode for categorical
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
```
